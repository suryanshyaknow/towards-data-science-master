```python
data4 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
df4 = pd.DataFrame(data4)
df4
```

```python
data5  = {'emp_id' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
         }
df5 = pd.DataFrame(data5)
df5
```

```python
pd.merge(df4, df5)
```

```python
### Left Join:

pd.merge(df4, df5, how='left')
```

```python
### Right Join:

pd.merge(df4, df5, how='right')
```

```python
### Outer Join: use union of keys from both frames

pd.merge(df4, df5, how='outer')
```

```python
df4
```

```python
df5
```

```python
### Cross-join: Creates the cartesian product from both frames, preserves the order of the left keys.

pd.merge(df4, df5, how='cross')
```

### If column names are different, than the interpreter won't be able to pick the column automatically on which merge is to be applied on.

```python
data6 = {'emp_id1' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
df6 = pd.DataFrame(data6)
         
data7  = {'emp_id2' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
         }
df7 = pd.DataFrame(data7)
```

```python
df6
```

```python
df7
```

```python
try:
    pd.merge(df6, df7)
except Exception as e:
    print(e)
```

```python
pd.merge(df6, df7, left_on='emp_id1', right_on='emp_id2')
```

```python
df_sales = pd.read_csv("sales_data_final.csv")
df_sales.head()
```

```python
def profit_flag(profit):
    if profit > 0:
        return 'positive'
    else:
        return 'negative'
```

```python
df_sales['profit_flag'] = df_sales['profit'].apply(profit_flag)

### Ain't no nned to pass parantheses while passing function in the apply method.
```

```python
df_sales.head()
```

```python
def quantity_flag(quantity):
    
    if quantity <= 10:
        return "Low"
    elif 10 < quantity <= 10:
        return "Medium"
    else:
        return "High"
```

```python
df_sales["quantity_flag"] = df_sales["quantity"].apply(quantity_flag)
```
