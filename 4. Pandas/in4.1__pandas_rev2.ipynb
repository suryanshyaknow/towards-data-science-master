```python
df2.columns
```

```python
df2.index
```

```python
rows_thresh = 0.3*9 
rows_thresh
```

```python
cols_thresh = .3*51290
cols_thresh
```

```python
## So if any column contains more than 15387 NaN vals drop it and the same goes for rows if it contains more than 3 NaN vals

df2_desired = df2.dropna(thresh = 3)
df2_desired_ = df2_desired.dropna(thresh = cols_thresh/3, axis=1)
```
