**According to the U.S. Department of Health and Human Services, the average adult woman expends roughly 1,600 to 2,400 calories per day, and the average adult man uses 2,000 to 3,000 calories per day.**

```python
# Converting the datatype of Id column into 'object' type 
df1['Id'] = df1['Id'].astype('object')
```

```python
### 8. Number of healthy persons (on the basis of Calories burnt)

# Since we don't know the gender, we'll go with the range(2000, 3000)

df1[['Id']].where((df1['Calories'] >= 2000) & (df1['Calories'] <= 3000)).dropna().nunique() # --> 30 healthy folks
```

```python
### 9. How many persons are not regular with respect to activity?

#--> Let's assume a criteria that if a person appears to be active 5 days a week for more than two weeks, he/she's regular.
# i.e if activity_days > 10
```

```python
df1.head(2)
```

```python
regularity = pd.DataFrame(df1[['Id']].drop_duplicates())
regularity.set_index('Id', inplace=True)

# creating the column 'ActiveDays'
regularity['ActiveDays'] = df1.groupby('Id')['ActivityDate'].count()

regularity.head()
```

```python
regularity.where(regularity.ActiveDays>10).dropna()
```

```python
### 10. Third most active person:
# active person: the one who has been active on most days

mostActive3 = df1.groupby('Id')[['ActivityDate']].count().sort_values(by=['ActivityDate'],ascending=False).head(3) 
# mostActive3: The three most active people sorted in decreasing order

mostActive3.iloc[2] # the third most active person: 1503960366
# or mostActive3.index[2]
```

```python
df1.head()
```

```python
### 11. The 5th most laziest person:
# laziest: who has walked least number of steps

laziest5 = df1.groupby('Id')[['TotalSteps']].sum().sort_values(by=['TotalSteps']).head(5)

fifth_laziest = laziest5.index[4] # top 5 laziest persons in ascending order
fifth_laziest
```

```python
### 12. Total Accumulative Calories burnt for each person

df1.groupby('Id')[['Calories']].sum()
```

```python
### 7. Find out how many different-different shipment modes that we have 
## and what is a percentage usability of all the shipment mode with respect to dataset.
```

```python
### 8. Create a new coulmn and try to find out the difference between order date and shipment date.
```

```python
# Converting the date columns of the deep copy of orders df into datetime format:

orders['Order Date'] = orders['Order Date'].astype('datetime64')
orders['Ship Date'] = orders['Ship Date'].astype('datetime64')
```

```python
# "the difference between order date and shipment date" column:

orders["Shipment Duration"] = orders['Ship Date'].dt.date - orders['Order Date'].dt.date
orders
```

```python
orders.where(orders['Shipment Duration'] == '84 days').dropna()
```

```python
### 9. Based on the above question, find out for which order ids we have shipment duration more than 10 days.
```

```python
# creating a column that contains the vals of "Shipment Duration" column into the int datatypes:

orders['Shipment Duration (int)'] = orders['Shipment Duration'].astype(str).apply(lambda x: int(x[:-5]))
```

```python
orders.head(2)
```

```python
# Order Ids having shipment duration > 10:

orders.where(orders['Shipment Duration (int)'] > 10).dropna()[['Order ID']].drop_duplicates().set_index('Order ID')
```

```python
### 10. Try to find out a list of a returned orders for which shipment duration was more than 15 days 
# and find out that region manager as well.
```

```python
# we found this in the 4th question

combined_w_returns = orders.join(returns.set_index('Order ID'), on='Order ID')
combined_w_returns.head()
```

```python
combined_w_returns[['Shipment Duration (int)', 'Status']].where((combined_w_returns['Shipment Duration (int)']>15) & (combined_w_returns["Status"] == 'returned')).dropna()
```

```python
### So this demonstrates that there are no order Ids that returned a product & of which shipment duration > 15 too.
```
