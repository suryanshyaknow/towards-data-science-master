```python
ds1=pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
ds1
```

```python
# Create a column having total price of items i.e. quantity * item_price:
```

```python
price = [float(ds1['item_price'][i][1:]) for i in range(len(ds1['item_price']))]
ds1['Cost paid'] = price * ds1.quantity
```

```python
ds1.drop(["cost"], axis=1, inplace=True)
```

```python
ds1
```

```python
# and we wanna add $ symbol in front of total cost:
ds1['Cost_paid'] = f"${ds1['Cost paid']}" #Ain't gonna work.
ds1.head()
```

```python
# that didn't work so we gotta use loop:
for i in range(len(ds1['Cost paid'])):
    ds1['Cost_paid'][i] = f"${ds1['Cost paid'][i]}"
```
