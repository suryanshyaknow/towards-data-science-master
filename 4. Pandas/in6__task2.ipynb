### 14. Total percentage of active hour for the month

```python
try:
    pd.to_datetime(df1['Duration'])
except Exception as e:
    print(e)
    
### It seems data is somewhat skewed
```

```python
df1.tail()

### look at the 999th val
```

```python
## dropiing the rows which aren't containing proper hour format

for i in range(len(df1)):
    if int(df1['Duration'][i].split(':')[0]) >= 24:
        df1.drop(i, inplace=True)
```

```python
### Active hour:

df1['active_hour'] = pd.to_datetime(df1['Duration']).dt.hour +\
                     pd.to_datetime(df1['Duration']).dt.minute/60 +\
                     pd.to_datetime(df1['Duration']).dt.second/3600
df1.head()
```

```python
### % active hour:

df1["% Active Hour"] = df1['active_hour']*100/df1['active_hour'].sum()
df1
```
